```python
hate = ann.loc[ann.label=='hate']
non_hate = ann.loc[ann.label=='noHate']
```

```python
non_hate.info()
```

```python
files = np.array(list(map(lambda x: x[0:-4], files)))
```

```python
files[:5]
```

```python
hh = hate[hate['file_id'].str.contains('12834217_4')]
if hh.any().any():
    print(hh['file_id'])
```

```python
texts = []
labels =[]

for file in files:
    with open(os.path.join(files_path,file+'.txt')) as f:
        sus = hate[hate['file_id'].str.contains(file)]
        if sus.any().any():
            texts.append(f.read())
            labels.append(1)
        elif non_hate[non_hate['file_id'].str.contains(file)].any().any():
            try:
                texts.append(f.read())
                labels.append(0)
            except:
                continue
```

```python
#cleaning data

import re
def clean_text(text):
  text = re.sub("RT",' ',text)
  text = re.sub("!{3,}",' ',text)
  text = re.sub(r"&#8217;", "'", text)
  text = re.sub(r"&[#*a-zA-z0-9]+;", '', text)
  text = re.sub("@[_*a-zA-Z0-9]+:*",' ',text)
  text = re.sub(r"https?://[A-Za-z0-9./]+", ' ', text)
  #text = re.sub(r"[^a-zA-z.!?'0-9]", ' ', text)
  text = re.sub('\t+', '',  text)
  text = re.sub(" {2,}",' ',text)
  text = re.sub("^ ",'',text)
  #text = re.sub(r" +", ' ', text)
  return text
```

```python
general['tweet'] = list(map(clean_text, general['tweet']))
```

```python
def listToString(s): 
    
    # initialize an empty string
    str1 = "" 
    
    # traverse in the string  
    for ele in s: 
        str1 += ele+" "  
    
    # return string  
    return str1 
racist2['post_tokens'] = racist2['post_tokens'].apply(listToString)
```

# Data Augmentation

```python

import nlpaug.augmenter.char as nac
import nlpaug.augmenter.word as naw
import nlpaug.augmenter.sentence as nas
import nlpaug.flow as nafc
```

```python
import nlpaug.augmenter.word as naw
from tqdm import tqdm
pure_isl = isl.loc[isl['class']==4,'tweet']
pure_isl = pure_isl.reset_index()
aug = naw.ContextualWordEmbsAug(
    model_path='bert-base-uncased', action="insert")
def augment(df, label, maxx):
  df = df['tweet']
  augmented = pd.DataFrame(columns=['class', 'tweet'])
  for i in tqdm(range(min(len(df), maxx))):
    text = df[i]
    augs = aug.augment(text)
    augmented = augmented.append({'class':label, 'tweet': augs}, ignore_index=True)
  return augmented
```

```python


pure_race = racist.loc[racist['class']==2,'tweet']
pure_race = pure_race.reset_index()
race_augmented = augment(pure_race, 2, 30)

pure_sexism = sexism.loc[sexism['class']==3,'tweet']
pure_sexism = pure_sexism.reset_index()
sexism_augmented = augment(pure_sexism, 3, 30)

pure_isl = isl.loc[isl['class']==4,'tweet']
pure_isl = pure_isl.reset_index()
race_augmented = augment(pure_isl, 4, 30)
```
